# Análise de Dados de Financiamentos.

```python
# Importação de bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Configuração de estilo para os gráficos

try:
    plt.style.use('seaborn-v0_8') # Tenta usar uma versão específica do estilo seaborn
except OSError:
    print("Estilo 'seaborn-v0_8' não encontrado. Usando o estilo padrão.")
    # Retorna ao estilo matplotlib padrão se os estilos seaborn não estiverem disponíveis
    plt.style.use('default')

%matplotlib inline

# Carregar o arquivo CSV
try:
    df = pd.read_csv('dados.csv', encoding='latin1')  # Usando latin1 para lidar com caracteres especiais
    print("Arquivo carregado com sucesso!")
except FileNotFoundError:
    print("Erro: O arquivo 'dados.csv' não foi encontrado. Por favor, verifique o caminho.")
    # Sai ou verifica se o arquivo dados.csv foi encontrado
    exit()
except Exception as e:
    print(f"Erro ao carregar o arquivo: {e}")
    exit()

# Visualização inicial dos dados
print("\nPrimeiras linhas do dataset:")
display(df.head())

print("\nInformações sobre o dataset:")
display(df.info())

print("\nEstatísticas descritivas:")
display(df.describe(include='all'))

# Limpeza e preparação dos dados
# Verificar valores nulos
print("\nValores nulos por coluna:")
display(df.isnull().sum())

# Converter DATA_AQUISICAO para datetime, tratando possíveis erros
try:
    df['DATA_AQUISICAO'] = pd.to_datetime(df['DATA_AQUISICAO'], dayfirst=True, errors='coerce')
    # Remove a conversão de data onde tem falha
    df.dropna(subset=['DATA_AQUISICAO'], inplace=True)
except KeyError:
    print("Erro: A coluna 'DATA_AQUISICAO' não foi encontrada no DataFrame.")
except Exception as e:
    print(f"Erro ao converter a coluna 'DATA_AQUISICAO': {e}")


# Análise por produto
if 'PRODUTO' in df.columns:
    print("\nContagem por tipo de produto:")
    produto_counts = df['PRODUTO'].value_counts()
    display(produto_counts)

    # Gráfico 1: Distribuição por tipo de produto
    if not produto_counts.empty:
        plt.figure(figsize=(10, 6))
        produto_counts.plot(kind='bar', color='skyblue')
        plt.title('Distribuição de Financiamentos por Tipo de Produto')
        plt.xlabel('Tipo de Produto')
        plt.ylabel('Quantidade')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
else:
    print("\nColuna 'PRODUTO' não encontrada. Pulando análises por produto.")

# Gráfico 2: Distribuição de idade dos clientes
if 'IDADE_CLIENTE' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df['IDADE_CLIENTE'].dropna(), bins=30, kde=True, color='green') # Dropna para lidar com NaNs
    plt.title('Distribuição de Idade dos Clientes')
    plt.xlabel('Idade do Cliente')
    plt.ylabel('Frequência')
    plt.tight_layout()
    plt.show()
else:
     print("\nColuna 'IDADE_CLIENTE' não encontrada. Pulando análise de idade.")

# Gráfico 3: Valor médio de tabela por produto
if 'PRODUTO' in df.columns and 'VALOR_TABELA' in df.columns:
    plt.figure(figsize=(10, 6))
    # Dropna para lidar com NaNs em ambas as colunas antes de agrupar
    df_clean = df.dropna(subset=['PRODUTO', 'VALOR_TABELA'])
    if not df_clean.empty:
        df_clean.groupby('PRODUTO')['VALOR_TABELA'].mean().sort_values().plot(kind='barh', color='orange')
        plt.title('Valor Médio de Tabela por Tipo de Produto')
        plt.xlabel('Valor Médio (R$)')
        plt.ylabel('Tipo de Produto')
        plt.tight_layout()
        plt.show()
else:
     print("\nColunas 'PRODUTO' ou 'VALOR_TABELA' não encontradas. Pulando análise de valor médio por produto.")

# Gráfico 4: Dias ativo vs. Parcelas pagas
if 'DIAS_ATIVO' in df.columns and 'QT_PC_PAGAS' in df.columns and 'PRODUTO' in df.columns:
    plt.figure(figsize=(10, 6))
    # Dropna para lidar com NaNs nas colunas usadas no scatterplot
    df_clean = df.dropna(subset=['DIAS_ATIVO', 'QT_PC_PAGAS', 'PRODUTO'])
    if not df_clean.empty:
        sns.scatterplot(data=df_clean, x='DIAS_ATIVO', y='QT_PC_PAGAS', hue='PRODUTO', alpha=0.7)
        plt.title('Relação entre Dias Ativo e Parcelas Pagas')
        plt.xlabel('Dias Ativo')
        plt.ylabel('Parcelas Pagas')
        plt.tight_layout()
        plt.show()
else:
     print("\nColunas 'DIAS_ATIVO', 'QT_PC_PAGAS' ou 'PRODUTO' não encontradas. Pulando scatterplot.")

# Gráfico 5: Distribuição de parcelas em atraso
if 'PRODUTO' in df.columns and 'QT_PC_VENCIDAS' in df.columns:
    plt.figure(figsize=(10, 6))
    # Dropna para lidar com NaNs nas colunas usadas no boxplot
    df_clean = df.dropna(subset=['PRODUTO', 'QT_PC_VENCIDAS'])
    if not df_clean.empty:
        sns.boxplot(data=df_clean, x='PRODUTO', y='QT_PC_VENCIDAS')
        plt.title('Distribuição de Parcelas em Atraso por Tipo de Produto')
        plt.xlabel('Tipo de Produto')
        plt.ylabel('Parcelas em Atraso')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
else:
     print("\nColunas 'PRODUTO' ou 'QT_PC_VENCIDAS' não encontradas. Pulando boxplot.")

# Gráfico 6: Evolução temporal de aquisições
if 'DATA_AQUISICAO' in df.columns:
    plt.figure(figsize=(12, 6))
    # Dropna para lidar com NaNs na coluna de data antes de contar e ordenar
    data_counts = df['DATA_AQUISICAO'].dropna().value_counts().sort_index()
    if not data_counts.empty:
        data_counts.plot()
        plt.title('Evolução Temporal de Aquisições')
        plt.xlabel('Data')
        plt.ylabel('Número de Aquisições')
        plt.tight_layout()
        plt.show()
else:
     print("\nColuna 'DATA_AQUISICAO' não encontrada. Pulando análise temporal.")

# Análise por faixa etária
if 'IDADE_CLIENTE' in df.columns:
    bins = [0, 30, 40, 50, 60, 70, 80, 90, 100]
    labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90+']
    # Garante que a coluna 'IDADE_CLIENTE' é numerica lida com os potenciais erros ou NaNs
    df['IDADE_CLIENTE'] = pd.to_numeric(df['IDADE_CLIENTE'], errors='coerce')
    df['FAIXA_ETARIA'] = pd.cut(df['IDADE_CLIENTE'], bins=bins, labels=labels, right=False, include_lowest=True)

    # Gráfico 7: Distribuição por faixa etária
    print("\nDistribuição por faixa etária:")
    faixa_etaria_counts = df['FAIXA_ETARIA'].value_counts().sort_index()
    display(faixa_etaria_counts)

    if not faixa_etaria_counts.empty:
        plt.figure(figsize=(10, 6))
        faixa_etaria_counts.plot(kind='bar', color='purple')
        plt.title('Distribuição de Clientes por Faixa Etária')
        plt.xlabel('Faixa Etária')
        plt.ylabel('Quantidade')
        plt.tight_layout()
        plt.show()

    # Gráfico 8: Valor médio de tabela por faixa etária
    if 'VALOR_TABELA' in df.columns:
         # Dropna para lidar com NaNs em ambas as colunas antes de agrupar
        df_clean = df.dropna(subset=['FAIXA_ETARIA', 'VALOR_TABELA'])
        if not df_clean.empty:
            plt.figure(figsize=(10, 6))
            df_clean.groupby('FAIXA_ETARIA')['VALOR_TABELA'].mean().plot(kind='bar', color='teal')
            plt.title('Valor Médio de Tabela por Faixa Etária')
            plt.xlabel('Faixa Etária')
            plt.ylabel('Valor Médio (R$)')
            plt.tight_layout()
            plt.show()
    else:
        print("\nColuna 'VALOR_TABELA' não encontrada. Pulando análise de valor médio por faixa etária.")
else:
    print("\nColuna 'IDADE_CLIENTE' não encontrada. Pulando análises por faixa etária.")


# Gráfico 9: Correlação entre variáveis numéricas
plt.figure(figsize=(10, 8))
# Selecione apenas colunas numéricas, manipulando NaNs potenciais antes da correlação
numeric_cols = df.select_dtypes(include=np.number).columns
if not numeric_cols.empty:
    # Calcular matriz de correlação, eliminando NaNs para cálculo de correlação
    corr_matrix = df[numeric_cols].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Matriz de Correlação entre Variáveis Numéricas')
    plt.tight_layout()
    plt.show()
else:
     print("\nNenhuma coluna numérica encontrada para calcular a correlação.")


# Gráfico 10: Distribuição de prazo de financiamento
if 'PRAZO_FINANCIAMENTO' in df.columns:
    plt.figure(figsize=(10, 6))
    # Dropna para lidar com NaNs na coluna de prazo antes de contar
    prazo_counts = df['PRAZO_FINANCIAMENTO'].dropna().value_counts()
    if not prazo_counts.empty:
        # Garante que os dados do gráfico de pizza sejam adequados
        if len(prazo_counts) > 20: # Examplo de limite
             print("\nMuitos valores únicos para 'PRAZO_FINANCIAMENTO'. Exibindo contagem.")
             display(prazo_counts)
        else:
            prazo_counts.plot(kind='pie', autopct='%1.1f%%')
            plt.title('Distribuição de Prazo de Financiamento (meses)')
            plt.ylabel('') # Remove o gráfico default e troca por pie(pizza)
            plt.tight_layout()
            plt.show()
else:
     print("\nColuna 'PRAZO_FINANCIAMENTO' não encontrada. Pulando gráfico de pizza.")
```
